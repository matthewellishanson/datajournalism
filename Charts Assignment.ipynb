```python
import agate
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(style="ticks")
%matplotlib inline
```

```python
mountainlions = agate.Table.from_csv('Downloads/mountainlions.csv')
print(mountainlions)
```

```python
print(mountainlions)
```

```python
by_county = mountainlions.group_by('COUNTY')
```

```python
county_totals = by_county.aggregate([('count', agate.Count())])
```

```python
county_totals.print_table()
```

```python
sorted_counties = county_totals.order_by('count', reverse=True)
sorted_counties.print_table(20)
```

```python
sorted_counties.to_csv('Downloads/mountainlions.csv')
```
